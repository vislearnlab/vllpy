```python
from vislearnlabpy.embeddings.stimuli_loader import ImageExtractor, ImgExtractionSettings
from dataclasses import replace

kisumu_extraction_settings_black = ImgExtractionSettings(apply_center_crop=True, crop_dim=1400, resize_dim=512, apply_content_crop=True, 
                                                        content_crop_padding=10, change_stroke_color=True, bg_threshold=1, 
                                                        bg_component_size=10, filter_edge_artifacts=True)

kisumu_extraction_settings_original_color = replace(
    kisumu_extraction_settings_black, 
    change_stroke_color=False,
    bg_threshold=200
)

ImageExtractor.save_transformed("test_kisumu.png", "kisumu_black.png", kisumu_extraction_settings_black)

ImageExtractor.save_transformed("test_kisumu.png", "kisumu_blue.png", kisumu_extraction_settings_original_color)
```
